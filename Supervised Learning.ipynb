```python
df = pd.read_csv('df.csv', index_col=0)
```

```python
numeric = ['Player','Age_Difference', 'Opponent', 'Height_Differential','Rank_Differential','Ranking_Points_Differential']
string = ['tourney_name','surface','Hand_Opposition']
```

```python
for col in df.columns:
    if col in numeric:
        # Standardisation des données numériques
        scaler = StandardScaler()
        df[col] = scaler.fit_transform(df[[col]])
    elif col in string:
        # Encodage des données de chaîne
        encoder = LabelEncoder()
        df[col] = encoder.fit_transform(df[col])
```

```python
df
```

```python
X=df.drop(columns=['Result'])
y=df['Result']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=123)
```

```python
rdf = RandomForestClassifier(n_estimators=300, max_depth=20,min_samples_split=3,min_samples_leaf=3,max_features='sqrt',random_state=42)
```

```python
rdf.fit(X_train,y_train)
```

```python
y_pred = rdf.predict(X_test)
```

```python
print('Classification report:\n', classification_report(y_test, y_pred))
```

```python
importances = rdf.feature_importances_
importances
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
# Extraction des éléments de la matrice de confusion
TP = conf_matrix[1, 1]  # Vrais positifs
FP = conf_matrix[0, 1]  # Faux positifs
TN = conf_matrix[0, 0]  # Vrais négatifs
FN = conf_matrix[1, 0]  # Faux négatifs

# Calcul de la précision et du rappel pour chaque classe
precision_positive = TP / (TP + FP)
recall_positive = TP / (TP + FN)

precision_negative = TN / (TN + FN)
recall_negative = TN / (TN + FP)
```

```python
print("Précision pour la classe positive :", precision_positive)
print("Rappel pour la classe positive :", recall_positive)

print("Précision pour la classe négative :", precision_negative)
print("Rappel pour la classe négative :", recall_negative)
```
